```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score as score
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
#from sklearn.tree import plot_tree
#import matplotlib.pyplot as plt
df = pd.read_csv("/content/drive/MyDrive/FDS2023-24/balance-scale.csv")
print(df)
df['Class'] = LabelEncoder().fit_transform(df['Class'].tolist())
y = df[['Class']]
X = df.drop(['Class'], axis = 1)

print(y.head())
print(X.head())
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 200)
clf = DecisionTreeClassifier()
clf.fit(train_X, train_y)
pred = clf.predict(test_X)
print('Accuracy in percent = ',score(pred, test_y)*100)
confusion_matrix = metrics.confusion_matrix(test_y, pred)
print('confusion matrix',confusion_matrix )
#plt.figure(figsize=(15, 10))
#plot_tree(clf, filled=True, feature_names=['L-Weight','L-Distance','R-Weight', 'R-Distance'], class_names=y, rounded=True)
#plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
#from sklearn.neighbors import KNeighborsClassifier
clf = RandomForestClassifier(n_estimators=1000, n_jobs=-1, )
clf.fit(train_X, train_y)
pred = clf.predict(test_X)
print('Accuracy in percent = ',score(pred, test_y)*100)
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(train_X, train_y)
pred = clf.predict(test_X)
print('Accuracy in percent = ',score(pred, test_y)*100)
```
