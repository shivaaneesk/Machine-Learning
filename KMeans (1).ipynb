```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Training Data creation
x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]
```

```python
data = list(zip(x, y))
print(data)
```

```python
# Test Data creation
x1 = [5, 2, 13]
y1 = [18,16,20]
data1 = list(zip(x1, y1))
print(data1)
```

```python
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)
print(kmeans.labels_)
pred_label= kmeans.predict(data1)
print("Predicted Value")
print(pred_label)
ss1= silhouette_score(data,kmeans.labels_,metric='euclidean')
ss= silhouette_score(data1,pred_label,metric='euclidean')
print("silhouette_score for training:",ss1)
print("silhouette_score for testing:",ss)
plt.scatter(x, y, c=kmeans.labels_)
plt.scatter(x1, y1, c=pred_label,edgecolors='red', linewidth=1)
plt.show()
```
